#### Starter Code for Question 2

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val = 0, left = None, right = None):
#         self.val = val
#         self.left = left
#         self.right = right

from typing import List

def bt_path(root: TreeNode) -> List[List[int]]:
    result = []

    def dfs(node, path):
        if not node:
            return
        
        # We add current node to the path
        path.append(node.val)

        # Then ,if leaf, we save the path
        if not node.left and not node.right:
            result.append(path[:])  # copy of current path
        else:
            dfs(node.left, path)
            dfs(node.right, path)

        path.pop()  # backtrack

    dfs(root, [])
    return result
```


-   Copy the solution your partner wrote. 


```python
# Your answer here
# Michelle's solution for Question 3

from typing import List

mylist = [1, 2, 0, 6, 0, 8]

def move_zeros_to_end(nums: List[int]) -> List[int]:
    write_index = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[write_index], nums[i] = nums[i], nums[write_index]
            write_index += 1
    return nums

move_zeros_to_end(mylist)
print(mylist)  # [1, 2, 6, 8, 0, 0]


```


-   Explain why their solution works in your own words.


```python
# Your answer here
Michelle’s code works by keeping a write_index that tracks where
the next non-zero number should go. It loops through the list, 
and whenever it finds a non-zero, it swaps it with the element at write_index and
then moves write_index forward. This way, 
all non-zero numbers stay in their original order, and zeros naturally get moved to the end.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Your answer here
Time Complexity: O(n)
The algorithm loops through the entire list once, checking each element.
As a result, the time complexity is linear,
proportional to the number of elements n in the list.

Space Complexity: O(1)
The algorithm uses a constant amount of extra space only a few variables are used 
(like write_index). Thus, the space complexity is constant, regardless of the list size.



```


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


```python
# Your answer here
Critique:
Michelle’s solution is correct and efficient, as it moves zeros to the end without altering the order of non-zero elements.
However, swapping elements when i == write_index is unnecessary and can be skipped for better efficiency.
 Additionally, while her time complexity is O(n),
the space complexity should be O(1) since the solution works in-place without using extra space.

Alternative solution:
Instead of swapping in-place, a simpler approach would be to create a new list with non-zero elements and append zeros at the end, 
though this would require extra space (O(n)).
```
